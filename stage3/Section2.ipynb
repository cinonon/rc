Section2：活性化関数

```python
import numpy as np
import matplotlib.pylab as plt

def relu(x):
    return np.maximum(0, x)

x = np.arange(-1, 1, 0.01)

y = relu(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```

```python

def sigmoid(x):
    a = 10
    return 1 / (1 + np.exp(-a * x))

x = np.arange(-1, 1, 0.01)

y = sigmoid(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```

```python

def step(x):
    return np.array(list(map(lambda  a: 1 if a>0 else 0, list(x))))

x = np.arange(-1, 1, 0.01)

y = step(x)
plt.plot(x, y)
plt.ylim(-0.1, 1.1)
```
